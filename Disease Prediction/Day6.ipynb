```python
#import the essential libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
diabetes_dataset = pd.read_csv('diabetes.csv')
```

```python
diabetes_dataset.head()
```

```python
diabetes_dataset.shape
```

```python
diabetes_dataset.describe()
```

```python
#check diabetees prsn 
diabetes_dataset['Outcome'].value_counts()
```

```python
# average of all colms
diabetes_dataset.groupby('Outcome').mean()
```

```python
diabetes_dataset.isnull().sum()
```

```python
# separation of the dataset into dependent and independent dataset
x = diabetes_dataset.drop(columns = 'Outcome', axis=1)
y = diabetes_dataset['Outcome']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
classifier = svm.SVC(kernel='linear')
classifier.fit(x_train,y_train)
```

```python
# accuracy score on the test data
y_pred = classifier.predict(x_test)
```

```python
test_data_accuracy = accuracy_score(y_pred,y_test)
print(test_data_accuracy)
```

```python
# prediction by the model on the random data
prediction = classifier.predict([[166,72,19,175,25.8,0.587,51]])[0]
print(prediction)
if (prediction == 0):
    print('The Person is not Diabetic')
else:
    print('The Person is Diabetic')
    
```

```python
# save data
import pickle
pickle.dump(classifier,open('diabetes_model.sav','wb'))
```

```python
#load data
model = pickle.load(open('diabetes_model.sav','rb'))
pred = model.predict([[166,72,19,175,25.8,0.587,51]])[0]

if (pred == 0):
    print('The Person is not Diabetic')
else:
    print('The Person is Diabetic')
```
